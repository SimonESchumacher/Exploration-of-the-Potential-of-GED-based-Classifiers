```python
from Test_protype_Selection import Base_Calculator,select_Prototype, Composite_Selection
from matplotlib import pyplot as plt
```

```python
calculator = Base_Calculator(size=800)
train_indexes = calculator.train_set
display(train_indexes)
```

```python
# display the train set
# trianing set is only index, the real elements is in the calculator.dataset
plt.figure(figsize=(6, 6))
plt.scatter(
    [calculator.dataset[i][0] for i in train_indexes],
    [calculator.dataset[i][1] for i in train_indexes],
    color='blue'
)
plt.title('Train Set')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.legend()
plt.show()
```

```python
protoypes = select_Prototype(G=train_indexes,ged_calculator=calculator,selection_method="k-CPS",size=20)
```

```python
display(protoypes)
```

```python
# display the coordinates of the prototypes
for i in protoypes:
    print(f"Prototype {i}: {calculator.dataset[i]}")
```

```python
# plot the prototypes
plt.figure(figsize=(8, 8))
train_without_prototypes = [i for i in train_indexes if i not in protoypes]
plt.scatter(    
    [calculator.dataset[i][0] for i in train_without_prototypes],
    [calculator.dataset[i][1] for i in train_without_prototypes],
    color='blue',alpha=0.5
)
plt.scatter(
    [calculator.dataset[i][0] for i in protoypes],
    [calculator.dataset[i][1] for i in protoypes],
    color='red'
)
plt.title('Prototypes Selection')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.legend()
plt.show()
```
